# 4874. Forth 

<br>

![image](https://user-images.githubusercontent.com/50367487/62832905-c5a82180-bc70-11e9-8f48-b6c1cd24e02a.png)

```
3
10 2 + 3 4 + * .
5 3 * + .
1 5 8 10 3 4 + + 3 + * 2 + + + .
```
```
#1 84
#2 error
#3 168
```

````python
TC = int(input())
for tc in range(TC):
    arr = []
    data = input().split()
    for i in data:
        if i.isdigit():
            arr.append(i)
        else:
            try:
                if i == '.': 
                    if len(arr) > 1:
                        print('#{} error'.format(tc + 1)); break
                    else:
                        print('#{} {}'.format(tc + 1, arr[0])); break
                b = int(arr.pop())
                a = int(arr.pop())
                if i == '+': c = a + b
                if i == '-': c = a - b
                if i == '*': c = a * b
                if i =='/':  c = a // b
                arr.append(c)
            except:
                print('#{} error'.format(tc + 1)); break
````

# 4880. 토너먼트 카드게임

```
#1 3
#2 5
#3 1
```

````python
import sys; sys.stdin = open('input.txt', 'r')

def win(l, r):
    if (arr[l], arr[r]) == (1, 3) or (arr[l], arr[r]) == (3, 2) or (arr[l], arr[r]) == (2, 1):
        return l
    elif (arr[l], arr[r]) == (3, 1) or (arr[l], arr[r]) ==  (2, 3) or (arr[l], arr[r]) == (1, 2):
        return r
    else: return l

def lr(s, f):
    if s == f: return s
    l = lr(s, (s + f)//2)
    r = lr((s + f)//2 + 1, f)
    return win(l, r)

for tc in range(int(input())):
    N = int(input())
    arr = list(map(int, input().split()))

    print('#{} {}'.format(tc + 1, lr(0, N - 1) + 1))
````
