````python
T = int(input())
for k in range(T):
    N = int(input())

    data = [[0]*10 for _ in range(10)]

    count = 0
    for _ in range(N):
        a, b, c, d, e = map(int, input().split())
        for i in range(c-a+1):
            for j in range(d-b+1):
                if data[a+i][b+j] == 0:
                    data[a+i][b+j] = e
                else:
                    if data[a+i][b+j] == e:
                        data[a+i][b+j] = e
                    else:
                        data[a+i][b+j] = e
                        count += 1
                    
    print('#{} {}'.format(k + 1, count))
````

# 4837. 부분집합의 합

![image](https://user-images.githubusercontent.com/50367487/62107558-cfca3780-b2e2-11e9-9bc1-b38770ce7333.png)

````python
T  = int(input())
data = list(range(1, 13))

for t in range(T):
    n , k = map(int, input().split())
    
    result = []
    for i in range(1 << len(data)):
        result.append([data[j] for j in range(len(data)) if (i & (1 << j))])
            
    count = 0
    for j in result:
        if len(j) == n and sum(j) == k:
            count += 1
            
    print(f'#{t+1} {count}')
````

# 4839. 이진탐색

![image](https://user-images.githubusercontent.com/50367487/62107598-e7a1bb80-b2e2-11e9-851d-9d68babf14a9.png)

````python
T = int(input())
for t in range(T):
    p, a, b = map(int, input().split())

    def cnt(p, x):
        l = 1
        count = 1
        while 1:
            c = (l + p )//2
            if c == x:
                return count
                break
            else:
                if c < x:
                    l = c
                else:
                    p = c
            count += 1
            
    if cnt(p, a) == cnt(p, b):
        print('#{} {}'.format(t + 1, 0) )
    else:
        if cnt(p, a) < cnt(p, b):
            print('#{} {}'.format(t + 1, 'A') )
        else:
            print('#{} {}'.format(t + 1, 'B') )
````

# 4843. 특별한 정렬



````python
TC = int(input())
for t in range(TC):
    N = int(input())
    data = sorted(map(int, input().split()))
    small, big = data[:N//2], data[N//2:]

    res = []
    while len(small + big) != 0:

        if big:
            res.append(big.pop())

        if small:
            res.append(small.pop(0)) 

    print('#{} {}'.format(t + 1, ' '.join(map(str, res))))
````
